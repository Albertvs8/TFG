## Analysis of the number of compositions in the comparison algorithm

```python
import numpy as np
import math
import matplotlib as mpl
import matplotlib.pyplot as plt

n=3
epsilon = [0.1,0.05,0.01,0.005,0.001,0.0005,0.0001,0.00005,0.00001,0.000001,0.0000001]
```

```python
def c_n(n):
    numerator = 2*n + 1
    denominator = 4**n
    binomial_coeff = math.comb(2*n, n)
    return numerator * binomial_coeff / denominator

c_4 = c_n(n)
c_4
```

```python
d=[]
df_dg=[]

def compute_d(n, cn, alpha, epsilon):
    log_c_n = np.log2(cn)
    log_epsilon = np.log2(1/epsilon)
    log_n_plus_1 = math.log2(n + 1)
    log_alpha_minus_2 = math.log2(alpha - 2)
    return (1/log_c_n) * log_epsilon + (1/log_n_plus_1) * log_alpha_minus_2


def compute_dg(cn, epsilon):
    log_c_n = np.log2(cn)
    log_epsilon = np.log2(1/epsilon)
    return (0.5/log_c_n) * log_epsilon

def compute_df(n, alpha):
    log_n = np.log2(n)
    log_alpha_minus_2 = math.log2(alpha - 2)
    return (1/log_n) * log_alpha_minus_2

alpha=10
print("e  |      d         |      dg        | df")
for e in epsilon:
    print(e,compute_d(n,c_4,alpha,e),compute_dg(c_4,e),compute_df(n,alpha))
    d.append(math.ceil(compute_d(n,c_4,alpha,e)))
    df_dg.append( math.ceil(compute_dg(c_4,e)) + math.ceil(compute_df(n,alpha)))
```

```python
log_epsilon = []

for eps in epsilon:
    log_eps = math.log2(eps)
    log_epsilon.append(-log_eps)

log_epsilon
```

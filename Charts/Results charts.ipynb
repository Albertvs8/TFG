# Experimental results

```python
import matplotlib.pyplot as plt
import numpy as np

def line_plot_individual(k,x_values, y_values_1,y_values_2,y_values_3,y_values_4):
    plt.plot(x_values, y_values_1, label='Encryption', linestyle='-', marker='o', color='blue')
    plt.plot(x_values, y_values_2, label='Distances', linestyle='--', marker='s', color='green')
    plt.plot(x_values, y_values_3, label='Comparisons', linestyle='-.', marker='^', color='red')
    plt.plot(x_values, y_values_4, label='Sorting', linestyle=':', marker='d', color='purple')


    plt.xlabel('Training samples')
    plt.ylabel('Time (s)')
    plt.title(f'k={k}')


    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.figure(figsize=(8, 6))
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def stacked_bar_plot(k, x_values, y_values_1, y_values_2, y_values_3, y_values_4):
    total = np.array([v1 + v2 + v3 + v4 for v1, v2, v3, v4 in zip(y_values_1, y_values_2, y_values_3, y_values_4)])
    percent1 = np.array([v1 / t * 100 for v1, t in zip(y_values_1, total)])
    percent2 = np.array([v2 / t * 100 for v2, t in zip(y_values_2, total)])
    percent3 = np.array([v3 / t * 100 for v3, t in zip(y_values_3, total)])
    percent4 = np.array([v4 / t * 100 for v4, t in zip(y_values_4, total)])

    plt.figure(figsize=(8, 6))

    colors = ['lightblue', '#8bc34a', '#ff5722', '#ff9800', '#9c27b0']

    plt.bar(x_values, percent1, label='Encryption', color=colors[0])
    plt.bar(x_values, percent2, bottom=percent1, label='Distances', color=colors[1])
    plt.bar(x_values, percent3, bottom=percent1 + percent2, label='Comparisons', color=colors[2])
    plt.bar(x_values, percent4, bottom=percent1 + percent2 + percent3, label='Sorting', color=colors[3])

    plt.xlabel('Training samples')
    plt.ylabel('Percentage')
    plt.title(f'k={k}')
    plt.ylim(0, 104)
    plt.legend(loc='upper right')


    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)


    plt.tick_params(axis='x', which='both', bottom=False)
    plt.tick_params(axis='y', which='both', left=False)


    plt.grid(axis='y', linestyle='--', alpha=0.7)

    if k == 3:
        show_percent_list = [0,1,2,3,4,5,6,7]
    elif k == 5:
        show_percent_list = [0,1,2,3,4,5]
    else:
        show_percent_list = [0,1,2,3]
        
    
    for i, v in enumerate(x_values):
        y_offset = 0
        for j, percent in enumerate([percent1, percent2, percent3, percent4]):
            if j in [1,2,3]:
                plt.text(v, y_offset + percent[i] / 2, f'{percent[i]:.1f}%', ha='center', va='center', fontsize=8,weight='bold',rotation=0)
            if i in show_percent_list and j == 0:
                 plt.text(v, y_offset + percent[i] / 2, f'{percent[i]:.1f}%', ha='center', va='center', fontsize=8,weight='bold',rotation=0)
                
            y_offset += percent[i]

    plt.show()
```
